#  Model Development
 <p>
    YOLO-NAS is an object detection algorithm developed by Deci AI, which is built using AutoNAC, a Neural Architecture Search Engine. It is designed to address the limitations of previous YOLO models and offers significant improvements in quantization support and accuracy-latency trade-offs.
  </p>

  <h2>Key Architectural Insights</h2>
  
  <ol>
    <li><strong>Quantization-Friendly Basic Block</strong>: YOLO-NAS introduces a new basic block that is friendly to quantization, addressing one of the significant limitations of previous YOLO models.</li>
    <li><strong>Sophisticated Training and Quantization</strong>: YOLO-NAS leverages advanced training schemes and post-training quantization to enhance performance.</li>
    <li><strong>QSP and QCI blocks</strong>: These blocks combine re-parameterization and 8-bit quantization advantages, allowing for minimal accuracy loss during post-training quantization.</li>
    <li><strong>AutoNAC</strong>: Deci's proprietary NAS technology was used to determine optimal sizes and structures of stages, including block type, number of blocks, and number of channels in each stage.</li>
  </ol>
  
  <h2>Model Variants</h2>
  
  <p>
    The YOLO-NAS models are primarily designed for object detection tasks and are available in small, medium, and large variants. The architecture is more efficient in terms of speed and memory, and it seamlessly supports inference engines like NVIDIA, making it a production-ready model.
  </p>

# **Methodology**
  <p>The YOLO-NAS object detection algorithm is a promising approach for automated kidney stone detection because it is able to achieve high accuracy and speed on a variety of object detection tasks.</p>

  <h2>Inside of YOLO-NAS</h2>
  <ul>
    <li>A new neural architecture search (NAS) algorithm specifically designed for object detection.</li>
    <li>A new training strategy combining supervised and unsupervised learning.</li>
    <li>A new post-processing algorithm to improve accuracy and speed.</li>
  </ul>

  <p>The YOLO-NAS algorithm has outperformed state-of-the-art object detection algorithms on various benchmarks, including the COCO dataset, and has proven effective for detecting kidney stones in CT scans.</p>

  <h2>YOLO-NAS Architecture</h2>
  <p>The YOLO-NAS architecture is a convolutional neural network (CNN) divided into two main components:</p>
  <ul>
    <li><strong>Backbone:</strong> A CNN feature extractor based on the RepVGG network.</li>
    <li><strong>Neck and Head:</strong> Responsible for object detection and localization, incorporating new attention mechanisms and post-processing algorithms.</li>
  </ul>

  <h2>How YOLO-NAS Works</h2>
  <p>The algorithm starts by feeding the input image into the backbone network, which extracts features for the neck and head networks. These latter networks then detect and localize the objects. The YOLO-NAS uses a single-stage detection approach, achieving both detection and localization in a single pass through the network, making it highly efficient.</p>

##  Training parameters



  <ul>
  <li><strong>max_epochs</strong> - Specifies the maximum number of training epochs. This controls how many times you'll work through the entire training dataset.</li>

  <li><strong>loss</strong> - The loss function you'll use for optimizing the model. It measures the model's performance on the dataset, and your aim is to minimize this function.</li>

  <li><strong>optimizer</strong> - The optimization algorithm you'll use to adjust the attributes of the neural network. Options include Adam, AdamW, SGD, Lion, or RMSProps.</li>

  <li><strong>train_metrics_list</strong> - A list of metrics to be logged during the training phase. These metrics help in monitoring the model's performance.</li>

  <li><strong>valid_metrics_list</strong> - A list of metrics to be logged during the validation phase. These metrics help you understand how well the model is generalizing.</li>

  <li><strong>metric_to_watch</strong> - This metric determines which model checkpoint will be saved. For example, you might choose to save the model that offers the highest accuracy on the validation set.</li>
  </ul>

```python
train_params = {
    # ENABLING SILENT MODE
    "average_best_models":True,
    "warmup_mode": "linear_epoch_step",
    "warmup_initial_lr": 1e-6,
    "lr_warmup_epochs": 3,
    "initial_lr": 5e-4,
    "lr_mode": "cosine",
    "cosine_final_lr_ratio": 0.1,
    "optimizer": "Adam",
    "optimizer_params": {"weight_decay": 0.0001},
    "zero_weight_decay_on_bias_and_bn": True,
    "ema": True,
    "ema_params": {"decay": 0.9, "decay_type": "threshold"},
    # ONLY TRAINING FOR 10 EPOCHS FOR THIS EXAMPLE NOTEBOOK
    "max_epochs": 50,
    "mixed_precision": True,
    "loss": PPYoloELoss(
        use_static_assigner=False,
        # NOTE: num_classes needs to be defined here
        num_classes=config.NUM_CLASSES,
        reg_max=16
    ),
    "valid_metrics_list": [
        DetectionMetrics_050(
            score_thres=0.1,
            top_k_predictions=300,
            # NOTE: num_classes needs to be defined here
            num_cls=config.NUM_CLASSES,
            normalize_targets=True,
            post_prediction_callback=PPYoloEPostPredictionCallback(
                score_threshold=0.01,
                nms_top_k=1000,
                max_predictions=300,
                nms_threshold=0.7
            )
        )
    ],
    "metric_to_watch": 'mAP@0.50'
}
```

```python
!pip install tqdm --upgrade
from tqdm.notebook import tqdm
import supervision as sv

### creating dataloader for test images
ds = sv.DetectionDataset.from_yolo(
    images_directory_path=f"/content/Kidney_Stone/test/images",
    annotations_directory_path=f"/content/Kidney_Stone/test/labels",
    data_yaml_path=f"/content/Kidney_Stone/data.yaml",
    force_masks=False
)
```

```python
CONFIDENCE_TRESHOLD = 0.5 ### setting up confidence value

predictions = {} ### prediction dictionary

for image_name, image in tqdm(ds.images.items()):
    result = list(best_model.predict(image, conf=CONFIDENCE_TRESHOLD))[0]
    detections = sv.Detections(
        xyxy=result.prediction.bboxes_xyxy,
        confidence=result.prediction.confidence,
        class_id=result.prediction.labels.astype(int)
    )
    predictions[image_name] = detections
```

```python
import random
random.seed(10)

MAX_IMAGE_COUNT = 5 ### number of images to be shown

n = min(MAX_IMAGE_COUNT, len(ds.images))

keys = list(ds.images.keys())
keys = random.sample(keys, n) ### randomly picking images

box_annotator = sv.BoxAnnotator() ### bbox annotator

images = []
titles = []

for key in keys:
    frame_with_annotations = box_annotator.annotate(
        scene=ds.images[key].copy(),
        detections=ds.annotations[key],
        skip_label=True
    )
    images.append(frame_with_annotations)
    titles.append('annotations')
    frame_with_predictions = box_annotator.annotate(
        scene=ds.images[key].copy(),
        detections=predictions[key],
        skip_label=True
    )
    images.append(frame_with_predictions)
    titles.append('predictions')

### plotting images with annotations and prediction values
%matplotlib inline
sv.plot_images_grid(images=images, titles=titles, grid_size=(n, 2), size=(2 * 4, n * 4))
```
